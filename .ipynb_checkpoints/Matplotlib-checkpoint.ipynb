### Scipy Assignment

```python
import pandas as pd
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
temp_max = np.array(Max)
temp_min = np.array(Min)
months = np.arange(12)
```

```python
def yearly_temps(times, avg, ampl, time_offset):
    return (avg+ ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      temp_max, [20, 10, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      temp_min, [-40, 20, 0])
```

```python
yearly_temps
```

```python
days = np.linspace(0, 12, num=365)

plt.figure(figsize=(8,9))
plt.plot(months, temp_max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, temp_min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

### Solution

```python
url=r'https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
titanic = pd.read_csv(url)
```

```python
titanic.head()
```

```python
gender=titanic.sex.value_counts()
```

```python
gender
```

```python
plt.figure(figsize=(8,10))
plt.pie(x=gender,labels=gender.index,explode=[0.01,0],autopct='%1.1f%%',shadow=True,startangle=140,pctdistance=0.5)
plt.legend(gender.index)
plt.title("Gender Proportion in Titanic data")
plt.show()
```

```python
titanic_male=titanic.loc[titanic.sex=='male']
titanic_female=titanic.loc[titanic.sex=='female']
```

```python
titanic_male.head()
```

```python
plt.figure(figsize=(10,12))
plt.scatter(data=titanic_male,x='age',y='fare',c='blue',s=30,label='male')
plt.scatter(data=titanic_female,x='age',y='fare',c='darkorange',s=30,label='female')
plt.legend()
plt.title("Fare distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Fare Paid")
plt.show()
```
